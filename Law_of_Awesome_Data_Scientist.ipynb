```python
import pandas as pd
```

```python
reserve_tb = pd.read_csv("/content/awesomebook/data/reserve.csv", encoding="UTF-8")
```

```python
reserve_tb.head()
```

```python
import sqlite3

with sqlite3.connect("tmp.sqlite3") as conn:
    reserve_tb.to_sql('reserve_tb', con=conn)
```

#### SQLによる前処理

```python
%%sql
SELECT
  -- reserve_idを選択(ASを利用して名前をrsv_timeに変更)
  reserve_id AS rsv_time,

  -- hotel_id, customer_id, reserve_datetimeを選択
  hotel_id, customer_id, reserve_datetime,

  -- checkin_data, checkin_time, checkout_dateを選択
  checkin_date, checkin_time, checkout_date

FROM reserve_tb
```

### 2-2 条件指定による抽出

#### SQLによる前処理

```python
%%sql
SELECT *
FROM reserve_tb

-- checkin_dateが、2016-10-12から2016-10-13までのデータに絞り込み
WHERE checkin_date BETWEEN '2016-10-12' AND '2016-10-13'
```

#### Pythonによる前処理

```python
reserve_tb.query('"2016-10-13" <= checkout_date <= "2016-10-14"')
```
